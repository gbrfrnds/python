```python
#IMPORTAÇÃO DE BIBLIOTECAS USADAS NO CÓDIGO
import pyautogui, os
import pandas as pd
from time import sleep
from selenium import webdriver
import time
import pyautogui
import pyperclip

arquivo = 'base.xlsx'
dados = pd.read_excel(arquivo)

fechar_msg_chrome = (1893, 87) #click para fechar a mensagem
nova_averbacao= (33, 131) #click para abrir nova averbação

#LOGIN SEFAZ
pyautogui.alert("Vai começar, aperte OK e não mexa em nada")

navegador = webdriver.Chrome()
navegador.get("https://tesouro.sefaz.rs.gov.br/Auth/Login.aspx")
time.sleep(2)
pyautogui.hotkey("alt", "space", "x")
pyautogui.press("x")

time.sleep(1)

pyautogui.moveTo(fechar_msg_chrome, duration=0.5)
pyautogui.click()

time.sleep(1)

navegador.find_element_by_xpath('//*[@id="txtUser"]').send_keys("usuario")
time.sleep(1)
navegador.find_element_by_xpath('//*[@id="txtPwd"]').send_keys("senha")
time.sleep(1)
navegador.find_element_by_xpath('//*[@id="btnLogin"]').click()

time.sleep(2)

pyautogui.moveTo(nova_averbacao, duration=0.5)
pyautogui.click()

time.sleep(3)

#posições de click
cmp_convenio = (835, 389) #click no convenio p/ abrir opções
cmp_conveniosel = (797, 452) #click para selecionar o gov ipe
cmp_servidor = (833, 417) #click no servidor, onde vai o ID
cmp_especieprin =(813, 442) #click na especie principal p/ abrir opções Point
cmp_especieprinsel = (802, 474) #click para selecionar emprestimo
cmp_cdaut = (682, 522) #click no cd. autenticação
cmp_ctr = (696, 550) #click no contrato 
cmp_valoraut = (628, 655) #click no valor autorizado
cmp_valini = (630, 679) #click na validade inicial
cmp_valfim = (776, 682) #click no validade final
cmp_salvar = (509, 295) #click em salvar
salvar_paint1 = (39, 9) #salvar paint icone
salvar_paint2 = (1489, 796) #salvar paint botao salvar
cmp_novo = (444, 295) #click no botao novo

#Preenchimento dos campos 
for i, servidor in enumerate(dados["SERVIDOR"]):
    cod_aut = dados.loc[i, "COD_AUT"]
    ctr = dados.loc[i, "CTR"]
    vlr_aut = dados.loc[i, "VLR_AUT"]
    val_ini = dados.loc[i, "VAL_INI"]
    val_fim = dados.loc[i, "VAL_FIM"]
    nome_print = dados.loc[i, "NOME_PRINT"]
    
    #Mover para posição e clicar
    pyautogui.moveTo(cmp_convenio, duration=0.5)
    pyautogui.click()
    pyautogui.moveTo(cmp_conveniosel, duration=0.5)
    pyautogui.click()
    pyautogui.moveTo(cmp_servidor, duration=0.5)
    pyautogui.click()
    
    #Aguardar e digitar o conteúdo da planilha

    sleep(1)

    def add_zero(servidor):
        if len(str(servidor))==11:
            return '0' + str(servidor)
        if len(str(servidor))==10:
            return '00' + str(servidor)
        if len(str(servidor))==9:
            return '000' + str(servidor)
        if len(str(servidor))==8:
            return '0000' + str(servidor)
        if len(str(servidor))==7:
            return '00000' + str(servidor)
        if len(str(servidor))==6:
            return '000000' + str(servidor)
        if len(str(servidor))==5:
            return '0000000' + str(servidor)
        
    pyautogui.typewrite(str(add_zero(servidor)))
    
    #Mover para posição e clicar
    sleep(2)
    pyautogui.moveTo(cmp_especieprin, duration=0.5)
    pyautogui.click()
    pyautogui.moveTo(cmp_especieprin, duration=1)
    pyautogui.click()
    pyautogui.moveTo(cmp_especieprinsel, duration=0.5)
    pyautogui.click()
    
    #Mover para posição e clicar cod aut
    sleep(1)
    pyautogui.moveTo(cmp_cdaut, duration=1)
    pyautogui.click()
    pyautogui.typewrite(str(cod_aut))
    
    #Mover para posição e clicar ctr
    sleep(1)
    pyautogui.moveTo(cmp_ctr, duration=1)
    pyautogui.click()
    pyautogui.typewrite(str(ctr))
    
    #Mover para posição e clicar Valor
    sleep(1)
    pyautogui.moveTo(cmp_valoraut, duration=1)
    pyautogui.click()
    pyautogui.typewrite(str(vlr_aut).replace(".",","))
    
    #Mover para posição e clicar Validade inicial
    sleep(1)
    pyautogui.moveTo(cmp_valini, duration=1)
    pyautogui.click()
    pyautogui.typewrite(str(val_ini))
    
    #Mover para posição e clicar Validade final
    sleep(1)
    pyautogui.moveTo(cmp_valfim, duration=1)
    pyautogui.click()
    pyautogui.typewrite(str(val_fim))
    
    #Mover para posição e clicar em Salvar
    sleep(1)
    pyautogui.moveTo(cmp_salvar, duration=1)
    pyautogui.click()
    
    sleep(7)
    pyautogui.press("prntscrn")
    
    #Abrir paint para salvar print
    sleep(2)
    pyautogui.press("winleft")
    sleep(1)
    pyautogui.write("paint")
    sleep(1)
    pyautogui.press("enter")
    sleep(1)
    pyautogui.hotkey('ctrl', 'v')
    sleep(1)
    pyautogui.moveTo(salvar_paint1, duration=1) #salvar paint
    pyautogui.click()
    sleep(1)
    pyautogui.typewrite(nome_print, interval=0.05)
    pyautogui.moveTo(salvar_paint2, duration=0.5)
    pyautogui.click()
    sleep(1)
    pyautogui.hotkey('alt', 'f4')
    
    #Mover para posição inicial em novo
    pyautogui.moveTo(cmp_novo, duration=1)
    pyautogui.click()

    sleep(5)
```
